```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Anscombe's quartet consists of four datasets that have nearly identical simple statistical properties (mean, standard deviation, correlation, and linear regression line) but look completely different when plotted.

```python
df= sns.load_dataset("anscombe")
df.head()
```

```python
df.dataset.unique()
```

```python
df[df.dataset=="II"].head()
```

```python
sns.lmplot(x="x", y="y", data=df.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=df.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```

```python
sns.lmplot(x="x", y="y", data=df.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```
