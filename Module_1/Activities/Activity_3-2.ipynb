#### 3. Data Visualization:

```python
gender = df['Gender'].map({"Male": 0, "Female": 1})
gender.value_counts().plot(kind= 'pie', colors=['blue', 'pink'], labels= ['Male', 'Female'], autopct='%1.1f%%', shadow = True, explode = [0,0.05])
plt.title("Gender Distribution")
plt.show()
```

#### a. Heatmap.

```python
#  Correlation matrix 
Correlation = Numerical_Val.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(Correlation, dtype=bool))

# Plot the heatmap with the mask
sns.heatmap(Correlation, cmap='coolwarm', linewidths=0.5, fmt='.2f', square=True, mask=mask)
plt.title('Correlation Heatmap (Lower Triangle)')
plt.show()
```
