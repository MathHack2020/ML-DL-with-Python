### I ) load dataset

```python
# Dataset Loading 
import pandas as pd
df = pd.read_csv("../data/Mkt_Banking.csv", index_col=[0])
```

### II) .check missing values

```python
# Counts all missing values across the entire DataFrame
df.isna().sum()
```

```python
# display rows having at least one missing value.
df[df.isna().any(axis=1)]
```

### III )  replace  all missing  in  "Age" , "contact". and  "Duration" 

```python
df.keys()
```

```python
df[["age", "duration"]].describe().T
```

```python
# Age mean , same as above desription
print(f"  Age Mean  : {df.age.mean()}")
```

```python
# reaplace missing value by the mean of all ages 
df.fillna({'age': df.age.mean()}, inplace=True)
```

### IV) replace missing values of the feature "duration" with the median value 

```python
# Duration  Median  (2nd Quartel ) ,
print(f"  Duration Median   : {df.duration.median()}")
```

```python
# reaplace missing value by the Median  of all duration   
df.fillna({'duration':df.duration.median()},inplace=True)
```

### IV) replace missing values of the feature "conctact " with the most frequent value 

```python
print(f"  Most frequent occurence    : {df.contact.mode()}")
```

```python
# reaplace missing value by most frequent occurnce   of  contact feature    
df.fillna({'contact':df.contact.mode()},inplace=True)
```
